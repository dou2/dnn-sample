```python
import keras
from keras import datasets
from keras.models import Model
from keras import Sequential,Input
from keras.layers import Dense
import numpy as np
import pandas as pd
from utils.all import *
from keras import backend as K
```

```python
X_train, Y_train, X_test, Y_test = load_dataset('d:/abc.csv')
m_samples,n_features = X_train.shape
```

```python
# 序贯模型
input_x = Input(shape=(n_features,))
d1 = Dense(20, activation='relu')(input_x)
d2 = Dense(10, activation='relu')(d1)
d3 = Dense(1, activation='sigmoid')(d2)
```

```python
model = Model(inputs=input_x, outputs=d3)
model.summary()
model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['acc'])
```

```python
history = model.fit(X_train, Y_train, batch_size=1000, epochs=100, validation_data=(X_test, Y_test))
```
